```python
data = pd.read_csv('input_files/day3_input.txt', dtype=object, header = None)
print(data.head())
```

### Part 1

```python
gamma_rate = ""
epsilon_rate = ""

for c in range(0, len(data.iloc[0,0])):
    count = 0
    for i in range(0, len(data)):
        count += float(data.iloc[i,0][c])

    if count > len(data)/2:
        gamma_rate = gamma_rate+"1"
    else:
        gamma_rate = gamma_rate+"0"
    
    print(count)
    print(gamma_rate)

print(gamma_rate)

for char in range(0, len(gamma_rate)):
    
    if gamma_rate[char] == "0":
        epsilon_rate = epsilon_rate + "1"
    else:
        epsilon_rate = epsilon_rate + "0"
        
print(epsilon_rate)


gamma_rate = int(gamma_rate, 2)
epsilon_rate = int(epsilon_rate, 2)
power = gamma_rate * epsilon_rate

print(gamma_rate)
print(epsilon_rate)
print(power)
```

### Part 2

```python
start_list = data[0].tolist()

for c in range(0, len(data.iloc[0,0])): #number of digits
    print("Checking char -----------------------", c)
    count = 0
    for i in range(0, len(start_list)):
        count += float(start_list[i][c])
    
    print("countof 1 %", count/len(start_list)*100)
    if count >= len(start_list)/2: ## = in case of parity give one
        most_common_value = "1"
    else:
        most_common_value = "0"

    print("Most common value", most_common_value)
    print("digits before change", [x[c] for x in start_list])
    print("len before change", len(start_list))

    if (len(start_list) > 2):
        start_list = [x for x in start_list if most_common_value in x[c]]
    else:
        start_list = [x for x in start_list if "1" in x[c]]
    
    print("digits after change", [x[c] for x in start_list])
    print("len after change", len(start_list))

    print(start_list)
        
    if len(start_list) == 1:
        break

start_list 
oxygen_generator_rating = start_list[0]
print("oxygen_generator_rating", oxygen_generator_rating,int(oxygen_generator_rating, 2))
```
