```python
def add_lists(list1, list2):
    print("adding", list1, list2)
    return "["+list1+","+list2+"]"

def find_explode(list):
    print("exploiding", list)
    prev_left, left_char, right_char, prev_right = None, None, None, None
    
    counter_opening = 0
    left_digit_found = False
    right_digit_found = False
    for i, char in enumerate(list):
        #print("index, char", i, char)
        is_pair = list[i-1]== "[" and char.isdigit() and list[i+1]=="," and list[i+2].isdigit() and list[i+3]=="]"
        #print(is_pair, list[i-1]+char+list[i+1]+list[i+2]+list[i+3])
        
        if char == "[":
            counter_opening+=1

        if counter_opening >= 5 and is_pair and left_digit_found==False and right_digit_found==False:
            left_char = char
            left_digit_found = True
            right_char = list[i+2]
            right_digit_found = True
            
            left_list_reversed = list[0:i][::-1]
            left_list = list[0:i]
            #print("left list - reversed", left_list_reversed, "normal left list", left_list)
            for index,subchar in enumerate(left_list_reversed):
                #print("left sub iteration", index, subchar)

                if subchar.isdigit():
                    prev_left = left_list_reversed[index:index+4][::-1]
                    #print("prev left found", prev_left)
                    break
            
            right_list = list[i+4:]
            #print("right list", right_list)
            for index, subchar in enumerate(right_list):
                #print("right sub iteration", index, subchar)
                if subchar.isdigit():
                    prev_right =right_list[index:index+4]
                    #print("prev right found", prev_right)
                    break

    prev_left = "" if prev_left is None else prev_left
    prev_right = "" if prev_right is None else prev_right
    
    if len(prev_left) >0:
        prev_left_replacement = prev_left[0:3]+str(int(prev_left[3])+int(left_char))
        list = list.replace(prev_left, prev_left_replacement)
    if len(prev_right) >0:
        prev_right_replacement = str(int(prev_right[0])+int(right_char))+prev_right[1:3]
        list = list.replace(prev_right, prev_right_replacement)
    
    print("found pair: ["+left_char+","+right_char+"]")
    list = list.replace("["+left_char+","+right_char+"]", "0")
    return list

import math

def find_split(list):
    print("splitting", list)
    for i, char in enumerate(list):
        #print(i, char)
        if i < len(list)-1:
            entire_nr = char+list[i+1]
            #print("entire", entire_nr)
            if entire_nr.isdigit() and int(entire_nr) >=10:
                left = str(math.floor(int(entire_nr)/2))
                right = str(math.ceil(int(entire_nr)/2))
                list = list.replace(entire_nr, "["+left+","+right+"]")
                print("replaced list", list)
    return list

def snailfish_addition(list1, list2):
    initial_run = add_lists(list1, list2)

    while initial_run != find_explode(initial_run):
        print("running explosion")
        initial_run = find_explode(initial_run)
        
    return initial_run
```

```python
import math
def find_pair_explode(list):
    #print("Checking", list)
    left_char, right_char, left_char_index, right_char_index = None, None, None, None
    
    counter_opening = 0
    left_digit_found = False
    right_digit_found = False
    for i, char in enumerate(list):
        is_pair = list[i-1]== "[" and char.isdigit() and list[i+1]=="," and list[i+2].isdigit() and list[i+3]=="]"
        #print(i, char, counter_opening)
        if counter_opening >= 5 and is_pair and left_digit_found==False and right_digit_found==False:
            left_char = char
            left_char_index = i
            left_digit_found = True
            right_char = list[i+2]
            right_char_index = i+2
            right_digit_found = True
            print("found pair: ["+left_char+","+right_char+"]", "counter", counter_opening)

        counter_opening += (1 if char == "[" else -1 if char == "]" else 0)

    if left_char == None and right_char == None:
        print("No pair found!")

    return left_char, right_char, left_char_index, right_char_index

def find_prev_post_numbers(list, left_char_index, right_char_index):
    prev_char, prev_char_index = "[", 0
    post_char, post_char_index = "]", len(list)
    
    for i in reversed(range(0, left_char_index)):
        #print(list[i])
        if list[i].isdigit():
            prev_char = list[i]
            prev_char_index = i
            #print("prev", prev_char, i)
            break
            
    for i in range(right_char_index+1, len(list)):
        if list[i].isdigit():
            post_char = list[i]
            post_char_index = i
            #print("post", post_char, i)
            break           
    return prev_char, post_char, prev_char_index, post_char_index

def add_lists(list1, list2):
    print("adding", list1, list2)
    return "["+list1+","+list2+"]"

def explode(list):
    print("exploding", list)
    left_char, right_char, left_char_index, right_char_index = find_pair_explode(list)
    
    if left_char !=None and right_char!=None:
        prev_char, post_char, prev_char_index, post_char_index = find_prev_post_numbers(list, left_char_index, right_char_index)

        replace_prev = "[" if prev_char == "[" else str(int(prev_char)+int(left_char))
        replace_post = "]" if post_char == "]" else str(int(post_char)+int(right_char))
        #print(left_char, right_char, left_char_index, right_char_index)
        #print(prev_char, post_char, prev_char_index, post_char_index)
        
        #print("here", list[0:prev_char_index], "-", replace_prev,"-",
        #      list[prev_char_index+1:left_char_index-1],"-", 0,
        #      list[right_char_index+2:post_char_index],
        #      replace_post+list[post_char_index+1:])
              
        list = (list[0:prev_char_index]+replace_prev+ 
               list[prev_char_index+1:left_char_index-1]  + "0" +
               list[right_char_index+2:post_char_index]+replace_post+list[post_char_index+1:])
    
    print("exploded", list)
    return list

def split(list):
    print("splitting", list)
    for i, char in enumerate(list):
        #print(i, char)
        if i < len(list)-1:
            entire_nr = char+list[i+1]
            #print("entire", entire_nr)
            if entire_nr.isdigit() and int(entire_nr) >=10:
                left = str(math.floor(int(entire_nr)/2))
                right = str(math.ceil(int(entire_nr)/2))
                list = list.replace(entire_nr, "["+left+","+right+"]")
                print("replaced list", list)
                break
    return list

def snailfish_addition(list1, list2):
    initial_run = None
    final_run = add_lists(list1, list2)

    while initial_run != final_run:
        initial_run = final_run
        final_run = explode(initial_run)
        if final_run == initial_run:
            initial_run = None
            while initial_run != final_run:
                initial_run = final_run
                final_run = split(initial_run)

    return initial_run
```

```python
def sum_all_numbers(lines):
    for i, number in enumerate(lines):
        if i == 0:
            prev_result = number
        else:
            print("summing", prev_result, "and", number)
            prev_result = snailfish_addition(prev_result, number)
            #if i == 2:
            #    break
    return prev_result

sum_all_numbers(lines)
```

```python
import math
import re
from typing import Union


class SnailfishNumber():
    def __init__(self, number: str):
        self.number = number

    def __add__(self, other : Union[str, 'SnailfishNumber']):
        if not isinstance(other, SnailfishNumber):
            return SnailfishNumber(f'[{self.number},{other}]')
        return SnailfishNumber(f'[{self.number},{other.number}]')

    def __str__(self):
        return self.number

    def __repr__(self):
        return f'sailfish_number: {self}'

    def find_next_explosion(self):
        open_bracket_count = 0
        for ix, letter in enumerate(self.number):
            if letter == '[':
                open_bracket_count += 1
            elif letter == ']':
                open_bracket_count -= 1
            if open_bracket_count == 5:
                exploding_pair = re.search(r'\[\d+,\d+]', self.number[ix:]).group()
                return (ix, exploding_pair)
        return (None, None)

    @staticmethod
    def get_number_from_pair(pair: str):
        return tuple(int(x) for x in re.findall(r'\d+', pair))

    @staticmethod
    def split_number(match):
        number_ = int(match.group())
        if number_ > 9:
            number_by_two = int(number_) / 2
            return f'[{math.floor(number_by_two)},{math.ceil(number_by_two)}]'
        return str(number_)

    @staticmethod
    def magnitude_of_pair(match):
        x, y = re.findall(r'\d+', match.group())
        return str(int(x)*3+int(y)*2)

    def explode(self):
        ix, pair = self.find_next_explosion()
        if not pair:
            return False
        (prev, following) = self.get_number_from_pair(pair)
        # split string along explosion pair
        number_prev = self.number[:ix]
        number_following = self.number[ix + len(pair):]
        # replace last integer
        number_prev = re.sub(r'(.*\D)(\d+)',
                             lambda m: m.group(1) + str(int(m.group(2)) + prev), number_prev, count=1)
        # replace first integer
        number_following = re.sub(r'\d+',
                                  lambda m: str(int(m.group()) + following), number_following, count=1)
        new_number = number_prev + '0' + number_following
        changed = new_number != self.number
        self.number = new_number
        return changed

    def split(self):
        #find all digits with a starting 1
        new_number = re.sub(r'[1-9]\d+', self.split_number, self.number, count=1)
        changed = new_number != self.number
        self.number = new_number
        return changed

    def reduce(self):
        while self.explode() or self.split():
            pass

    def magnitude(self):
        new_number = self.number
        while True:
            new_number = re.sub(r'\[\d+,\d+]', self.magnitude_of_pair, new_number)
            #break if all chars are numbers
            if re.search(r'^\d+$', new_number):
                break
        return new_number


if __name__ == '__main__':
    with open('input_files/day18_input.txt', 'r') as file:
        numbers_to_add = file.read().strip().split()

    #Part 1
    s_number = SnailfishNumber(numbers_to_add[0])
    for number in numbers_to_add[1:]:
        s_number += number
        s_number.reduce()

    print(s_number.magnitude())

    #Part 2
    max_magnitude = 0
    for left in numbers_to_add:
        for right in numbers_to_add:
            a = SnailfishNumber(left)+right
            a.reduce()
            max_magnitude = max(max_magnitude, int(a.magnitude()))

    print(max_magnitude)
```
