```python
def padding_input(input, size_padding):
    len_row = len(input[0])
    new_row = ["."*(len_row+size_padding*2)]
    top_bottom = new_row*size_padding
    middle = input.copy()
    for i, val in enumerate(input):
        middle[i]="."*size_padding+val+"."*size_padding
    
    padded_image = top_bottom + middle + top_bottom
    return padded_image 

def find_output_pixel(input, algorithm, x, y):
    string_adjacents = "".join([line[y-1:y+2] for line in input[x-1:x+2]])
    final_number = int("".join(["0" if s == "." else "1" for s in string_adjacents]), 2)
    output_pixel = "".join(algorithm)[final_number]
    #print(string_adjacents, final_number, output_pixel)
    return output_pixel

def create_empty_output(rows, columns):
    empty_output = ["."*(columns)]*rows
    return empty_output

def image_enhancement(input, algorithm):
    output = create_empty_output(len(input), len(input))
    for x in range(1, len(input)):
        for y in range(1, len(input)):
            output[x] = update_pixel(output[x], find_output_pixel(input, algorithm, x, y), y)
    return output

def run_double_enhancement(input, algorithm):
    output1 = image_enhancement(input, algorithm)
    print("output1",*output1,sep='\n')
    output2 = image_enhancement(output1, algorithm)
    print("output2",*output2,sep='\n')
    output_cut = [x[2:len(x)-2] for x in output2[2:-2]]
    return ''.join(output_cut).count("#")
    
def update_pixel(string, replacement, index):
    return string[:index] + replacement + string[index+1:]

def run_enhancement_ntimes(input_pic, algorithm, n):
    i = 0
    output = input_pic
    while i < n:
        output = image_enhancement(output, algorithm)
        i+=1
    print("output ",i,*output,sep='\n')

    #output_cut = [x[2:len(x)-2] for x in output2[2:-2]]
    return ''.join(output).count("#")
```

```python
def build_map(map, filler):
    w = len(map[0]) + 2
    new_map = []
    new_map.append([filler] * w)
    for line in map:
        new_map.append([filler] + line + [filler])
    new_map.append([filler] * w)
    return new_map


def get_neighbors(r, c, map, filler):
    bs = ""
    for rm in range(-1, 2):
        for cm in range(-1, 2):
            if r + rm < 0 or r + rm >= len(map) or c + cm < 0 or c + cm >= len(map[0]):
                bs += str(filler)
            else:
                bs += str(map[r + rm][c + cm])
    return 1 if alg[int(bs, 2)] == "#" else 0


data = open("input_files/day20_input.txt").read().strip().split("\n\n")
alg = data.pop(0)
map = [[0 if j == "." else 1 for j in i] for i in data[0].split("\n")]

for i in range(1, 50 + 1):
    # determine state of infinite pixels
    filler = 1 if alg[0] == "#" and alg[-1] == "." and not i % 2 else 0
    map = build_map(map, filler)
    # iterate over map and find updates
    changes = {}
    for r in range(len(map)):
        for c in range(len(map[0])):
            changes[(r, c)] = get_neighbors(r, c, map, filler)
    # apply updates to map
    for r, c in changes:
        map[r][c] = changes[(r, c)]
    if i == 2:
        print(f"Part 1: {sum([val for sublist in map for val in sublist])}")
print(f"Part 2: {sum([val for sublist in map for val in sublist])}")
```
