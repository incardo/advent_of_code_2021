### Part 1

```python
def update_position_step(S, initial_v):
    initial_x, final_x = S[0], S[0]
    initial_y, final_y = S[1], S[1]
    x_increase = initial_v[0]
    y_increase = initial_v[1]

    if initial_x == 0:
        final_x += x_increase
        new_x_increase = x_increase
    else:
        if x_increase > 0:
            final_x += x_increase-1
            new_x_increase = x_increase-1
        elif x_increase < 0:
            final_x += x_increase+1
            new_x_increase = x_increase+1
        else:
            final_x+= x_increase
            new_x_increase = x_increase
        
    if initial_y == 0:
        final_y += y_increase
        new_y_increase = y_increase-1
    else:
        final_y += y_increase
        new_y_increase = y_increase-1
        
    final_position = [final_x, final_y]
    final_speed = [new_x_increase, new_y_increase]
    #print("New position", final_position, final_speed)
    return final_position, final_speed

def run_n_steps(S, initial_v, n_steps, target_area):
    start = S
    speed = initial_v
    for _ in range(n_steps):
        start, speed = update_position_step(start, speed)
        check_target_area(start, target_area)
    return start, speed

def check_target_area(S, target_area):
    x, y = S[0], S[1]
    #print("checking:", S, target_area)
    xmin, xmax, ymin, ymax = target_area[0][0], target_area[0][1], target_area[1][0], target_area[1][1]
    if (xmin<=x<=xmax) and (ymin<=y<=ymax):
        #print("In target!")
        return True
    else:
        #print("Not in target")
        return False

def run_until_in_target(S, initial_v, target_area):
    start = S
    speed = initial_v
    xmin, xmax, ymin, ymax = target_area[0][0], target_area[0][1], target_area[1][0], target_area[1][1]
    max_y = S[1]
    
    target_found = check_target_area(start, target_area)
    steps = 0
    while target_found==False:
        start, speed = update_position_step(start, speed)
        target_found = check_target_area(start, target_area)
        x, y = start[0], start[1]
        max_y= y if y > max_y else max_y
        steps +=1
        if (x > xmax) or (y<ymin):
            #print("Target not achievable!!")
            break
            output = False

    output = True if target_found==True else False     
    #print("Check finished after ", steps, " steps - max y:", max_y, "Output", output)
    return output, start, speed, max_y
```

```python
S = [0,0]
initial_v = [6,9]
run_until_in_target(S, initial_v, target_area)
```

```python
max_y_reached = 0
for x in range(0, xmin):
    for y in range(0, xmin):
        #print("testing speed", x, y)
        output, start, speed, max_y = run_until_in_target(S, [x,y], target_area)
        max_y_reached= max_y if max_y > max_y_reached and output==True else max_y_reached

print("Max height reached:", max_y_reached)
```

```python
list_velocities = []
for x in range(-xmax-10, xmax+10):
    for y in range(-xmax-10, xmax+10):
        #print("testing speed", x, y)
        output, start, speed, max_y = run_until_in_target(S, [x,y], target_area)
        if output == True:
            list_velocities.append([x,y])
            
print("Length of correct velocities :", len(list_velocities))
print(list_velocities)
```
